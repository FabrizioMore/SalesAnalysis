#### Importar librerías necesarias

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
from itertools import combinations
from collections import Counter
```

### 1. Unir los 12 meses de datos en un solo archivo 

```python
files = [year for year in os.listdir('Sales_Data')]

all_months_data = pd.DataFrame()

for year in files:
    df = pd.read_csv("Sales_Data\\" + year)
    all_months_data = pd.concat([all_months_data, df])


all_months_data.to_csv("all_data.csv", index=False)
```

##### Leemos el DataFrame actualizado 

```python
all_data = pd.read_csv('all_data.csv')
all_data.head()
```

## Clean up the data

##### Eliminamos los erows con NaN

```python
nan_df = all_data[all_data.isna().any(axis=1)]


all_data = all_data.dropna(how='all')
all_data.head()
```

##### Eliminamos rows que contenian "Or"

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
all_data.head()
```

##### Convertimos las columnas "Price Each" y "Quantity Ordered" a int/float

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

## Agregamos columnas al DataFrame

### 2. Columna de mes

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

### 3. Columna ventas

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
```

```python
all_data.head()
```

### Pregunta 1: Cuál fué el mejor mes en ventas? Cuanto se ganó ese mes?

```python
results = all_data.groupby('Month').sum()
```

```python
months = range(1,13)

plt.bar(months, results['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month number')
plt.show()
```

##### Respuesta: El mejor mes fué Diciembre con más de 4 millonesde dólares en ventas

## 4. Añadir una columna de 'Ciudad'

```python
# Para esto, necesitamos .apply():

def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2][1:3]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: f'{get_city(x)} ({get_state(x)})')


all_data.head()
```

### Pregunta 2: Cuál fué la ciudad con el mayor número de ventas?

```python
results = all_data.groupby('City').sum()
results
```

```python
cities = [city for city, df in all_data.groupby('City')]

plt.bar(cities, results['Sales'])
plt.xticks(cities, rotation=45, size=8)
plt.xlabel('City name')
plt.ylabel('Sales in USD ($)')
plt.show()
```

##### Respuesta: La ciudad con mayor cantidad de ventas es San Francisco (CA)

### Pregunta 3: A qué hora deberíamos mostrar anuncios para maximizar la posibilidad de que los clientes compren los productos?

```python
all_data.head()
```

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute
```

```python
all_data.head()
```

```python
hours = [hour for hour, df in all_data.groupby(['Hour'])]

plt.plot(hours, all_data.groupby(['Hour']).count())
plt.xticks(hours)
plt.xlabel('Hour')
plt.ylabel('Number of Orders')
plt.grid()
```

##### Respuesta: Los mejores horarios son las 12 am y las 19 pm

### Pregunta 4: Qué productos suelen venderse juntos más a menudo?

```python
all_data.head()
```

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)]

df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

df = df[['Order ID', 'Grouped']].drop_duplicates()

df.head()
```

```python
count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key, value in count.most_common(10):
    print(key, value)
```

### Pregunta 5: Cuál producto vendió más? Por qué?

```python
all_data.head()
```

```python
product_group = all_data.groupby('Product')

sales_per_product = product_group['Sales'].sum()
```

```python
product_group = all_data.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']

products = [product for product, df in product_group]

plt.bar(products, quantity_ordered)
plt.xticks(products, rotation='vertical', size=8)
plt.xlabel('Product')
plt.ylabel('Quantity Ordered')
```

##### Respuesta: El producto más vendido es 'AAA Batteries (4-pack)'

```python
prices = all_data.groupby('Product').mean()['Price Each']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color='g')
ax2.plot(products, prices, 'b-')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color= 'g')
ax2.set_ylabel('Price in USD ($)', color= 'b')
ax1.set_xticklabels(products, rotation='vertical', size=8)

plt.show()
```
